```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8m-cls.pt")  # build a new model from scratch

# Use the model
data_path = r'/kaggle/input/emotion-detection-class'
model.train(data=data_path, epochs=200,patience=0, imgsz=640)

```

```python
model = YOLO('/kaggle/working/runs/classify/train5/weights/best.pt')

from PIL import Image
import cv2
import numpy as np
img = cv2.imread('/kaggle/input/emotion-detection-class/test/disgust/Z-1-_face_png_jpg.rf.80315f016eb39d31af4cffa8f909a12a.jpg')
results = model(img)
names_dict = results[0].names
probs = results[0].probs.tolist()
print (names_dict)
print (probs)
print(names_dict[np.argmax(probs)])
```
